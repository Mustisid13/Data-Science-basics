## Two Interfaces for the Price of One
### MATLAB-style interface

```python
plt.figure() #create a plot figure

# create first of two panel
plt.subplot(2,1,1) #(rows,columns,panel number)
plt.plot(x,np.sin(x))


#create second pannel and set current axis
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

### Object-oriented interface


```python
# First create a gric of plots
# ax will be an array of two Axes objects
fig,ax = plt.subplots(2)

ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

### Adjusting the Plot: Line Colors and Styles


```python
# different style in which you can assign colors to lines

plt.plot(x,np.sin(x),color='blue') #specify color by name
plt.plot(x,np.sin(x-1),color='g') #short color code(rbgcmyk)
plt.plot(x,np.sin(x-2),color='0.75') #greyscale
plt.plot(x,np.sin(x-3),color='#FFDD44') #hex code
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) #RGB tuple
```

```python
# different styl;e of lines 

plt.plot(x,np.sin(x),linestyle='-') #solid
plt.plot(x,np.sin(x-1),linestyle='--') #dashed
plt.plot(x,np.sin(x-2),linestyle='-.') #dashdot
plt.plot(x,np.sin(x-3),linestyle=':') #dotted
```

```python
# combine both 

plt.plot(x,x+1,'-g') #solid green
plt.plot(x,x+2,'--c') #dashed cyan
plt.plot(x,x+3,'-.k') #dashdot black
plt.plot(x,x+4,':r') #dotted red
```

```python
# plotting using plt.scatter
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
cmap='viridis')
plt.colorbar(); # show color scale
```

size(argument 's') adds up another feature to the graph, as we can see in below graph of irisdataset, third feature is added as size. So size of cirles plotted on graph will depend on the third feature, larger the value larger is the circle. Using argument 'c' we differentiated all three classes 

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

# Visualizing Errors

```python
# error bars
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
# continous errors

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
# Compute the Gaussian process fit
gp = RandomForestRegressor()
gp.fit(xdata[:, np.newaxis], ydata)
xfit = np.linspace(0, 10, 1000)
yfit = gp.predict(xfit[:, np.newaxis])
dyfit = 2 * np.sqrt(mean_squared_error(model(xfit),yfit)) # 2*sigma ~ 95% confidence region
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
color='gray', alpha=0.2)
plt.xlim(0, 10);
```

# Density and Contour Plots

```python
## Visualizing a 3D funcction
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contourf(X,Y,Z,cmap='RdGy')
plt.colorbar()
```

```python
# plotting as an image
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
cmap='RdGy')
plt.colorbar()
# plt.axis();
```

## 2D histogram

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x,y, bins=(30,30), cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

### plt.hexbin: Hexagonal binnings
