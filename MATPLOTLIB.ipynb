## Two Interfaces for the Price of One
### MATLAB-style interface

```python
plt.figure() #create a plot figure

# create first of two panel
plt.subplot(2,1,1) #(rows,columns,panel number)
plt.plot(x,np.sin(x))


#create second pannel and set current axis
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

### Object-oriented interface


```python
# First create a gric of plots
# ax will be an array of two Axes objects
fig,ax = plt.subplots(2)

ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

## Simple Line Plots

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x));
```

```python
# plot multiple line graph in one figure
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

### Adjusting the Plot: Line Colors and Styles


```python
# different style in which you can assign colors to lines

plt.plot(x,np.sin(x),color='blue') #specify color by name
plt.plot(x,np.sin(x-1),color='g') #short color code(rbgcmyk)
plt.plot(x,np.sin(x-2),color='0.75') #greyscale
plt.plot(x,np.sin(x-3),color='#FFDD44') #hex code
plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3)) #RGB tuple
```

```python
# different styl;e of lines 

plt.plot(x,np.sin(x),linestyle='-') #solid
plt.plot(x,np.sin(x-1),linestyle='--') #dashed
plt.plot(x,np.sin(x-2),linestyle='-.') #dashdot
plt.plot(x,np.sin(x-3),linestyle=':') #dotted
```

```python
# combine both 

plt.plot(x,x+1,'-g') #solid green
plt.plot(x,x+2,'--c') #dashed cyan
plt.plot(x,x+3,'-.k') #dashdot black
plt.plot(x,x+4,':r') #dotted red
```

### Adjusting the Plot: Axes Limits

```python
plt.plot(x,np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
#  The plt.axis() method allows you to set theand y limits with a single call, 
# by passing a list that specifies [xmin, xmax, ymin,ymax]

plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
# automatically tighten the bounds around the current plot

plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
#  ensuring an equal aspect ratiothat on your screen, one unit in x is equal to one unit in y
plt.plot(x, np.sin(x))
plt.axis('equal');
```

### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

## Matplotlib Gotchas



```python
# In the object-oriented interface to plotting, rather than calling these functions individually, 
# it is often more convenient to use the ax.set() method to set all these properties at once

ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sin(x)',title='A simple plot');
```
