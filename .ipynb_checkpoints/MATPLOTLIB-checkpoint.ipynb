# Visualizing Errors

```python
# error bars
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
# continous errors

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
# Compute the random forest regression process fit
gp = RandomForestRegressor()
gp.fit(xdata[:, np.newaxis], ydata)
xfit = np.linspace(0, 10, 1000)
yfit = gp.predict(xfit[:, np.newaxis])
dyfit = 2 * np.sqrt(mean_squared_error(model(xfit),yfit)) # 2*sigma ~ 95% confidence region
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
color='gray', alpha=0.2)
plt.xlim(0, 10);
```

# Density and Contour Plots

```python
## Visualizing a 3D function
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contourf(X,Y,Z,cmap='RdGy')
plt.colorbar()
```

```python
# plotting as an image
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
cmap='RdGy')
plt.colorbar()
# plt.axis();
```

```python
births = pd.read_csv('births.csv')

quartiles = np.percentile(births['births'],[25,50,75])
mu,sig = quartiles[1], 0.74 * (quartiles[2]-quartiles[0])
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births['day'] = births['day'].astype(int)
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, 
                              format='%Y%m%d')
births_by_date = births.pivot_table('births',[births.index.month,births.index.day])
births_by_date.index = [pd.datetime(2012,month,day) for (month,day) in births_by_date.index]
```

# Customizing Ticks
### Major and Minor Ticks
Within each axis, there is the concept of a major tick mark and a minor tick mark. As
the names would imply, major ticks are usually bigger or more pronounced, while
minor ticks are usually smaller. By default, Matplotlib rarely makes use of minor
ticks, but one place you can see them is within logarithmic plots

### Hiding Ticks or Labels

```python
ax = plt.axes()
ax.plot(np.random.rand(50))
ax.yaxis.set_major_locator(plt.NullLocator()) # locator formates only lines that shows the ticks
ax.xaxis.set_major_formatter(plt.NullFormatter()) # using formatter we only formate labels on the ticks
```

```python
# it can be usefuyl in displaying or plotting images
fig, ax = plt.subplots(5, 5, figsize=(5, 5))
fig.subplots_adjust(hspace=0, wspace=0)
# Get some face data from scikit-learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap="bone")
```

### Fancy Tick Formats

```python
# Plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')
# Set up grid, legend, and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi);
```

There are a couple changes we might like to make. First, its more natural for this data
to space the ticks and grid lines in multiples of  . We can do this by setting a Multi
pleLocator, which locates ticks at a multiple of the number you provide.

```python
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi/4))
fig
```

But now these tick labels look a little bit silly: we can see that they are multiples of ,
but the decimal representation does not immediately convey this.

```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
# "$\pi$" is rendered as the Greek character .
```

# Customizing Matplotlib: Configurations and Stylesheets


```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
# use a gray background
ax = plt.axes()
ax.set_axisbelow(True)
# draw solid white grid lines
plt.grid(color='w', linestyle='solid')
# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)
# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```
